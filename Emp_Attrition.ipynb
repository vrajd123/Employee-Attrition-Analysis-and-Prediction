```python
#Attrition with Department
plt.figure(figsize=(10,5))
sns.countplot(x='Department', hue='Attrition', data=data, palette='bright')
plt.title("Attrition with Department")
plt.show()
```

```python
#Attrition with EducationField
plt.figure(figsize=(10,5))
sns.countplot(x='EducationField', hue='Attrition', data=data, palette='bright')
plt.title("Attrition with EducationField")
plt.show()
```

```python
#Attrition with Jobrole
plt.figure(figsize=(10,5))
sns.countplot(x='JobRole', hue='Attrition', data=data, palette='bright')
plt.title("Attrition with Jobrole")
plt.xticks(rotation=45)
plt.show()
```

```python
# distribution of age
plt.figure(figsize=(10,5))
sns.distplot(data['Age'],hist=False)
plt.show()
```

```python
Numerical_col = ['Education','EnvironmentSatisfaction','JobInvolvement','JobSatisfaction',
                    'PerformanceRating','RelationshipSatisfaction','WorkLifeBalance']
data[Numerical_col].head()
```

```python
education_map = {1 :'Below College', 2: 'College', 3 :'Bachelor', 4 :'Master', 5: 'Doctor'}
plt.figure(figsize=(12,5))
sns.countplot(x=data['Education'].map(education_map), hue='Attrition', data=data, palette='hot')
plt.title("Education with Attrition")
plt.show()
```

```python
# Target Variable(Attrition)
data['Attrition'] = data['Attrition'].replace({'No':0,'Yes':1})
```

```python
#encoding binary variables
data['OverTime'] = data['OverTime'].map({'No':0,'Yes':1})
data['Gender'] = data['Gender'].map({'Male':0,'Female':1})
```
